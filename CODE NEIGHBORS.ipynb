```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
import pandas as pd
import pickle
```

```python
df = pd.read_csv('C:/Users/peni_/Desktop/proyecto/Proyecto-Final-Bootcamp---Dream-Team/data/df_processed_ML.csv')
df_cleaned = pd.read_csv('C:/Users/peni_/Desktop/proyecto/Proyecto-Final-Bootcamp---Dream-Team/data/df_final_cleaned.csv')
```

```python
df['url'] = df_cleaned['urls']
```

```python
df = df[['url'] + [col for col in df.columns if col != 'url']]
```

```python
df.info()
```

```python
df = df.drop(columns=['check_in_hour', 'check_out_hour', 'total_hours_checkin', 'log_num_reviews' ])
```

Seleccionar columnas relevantes para calcular similitud he usado los 11 primeros del feature importance

```python
features = [
    'prices_per_night', 'habitacion', 'baños', 'ratings', 'cleaning_fee',
    'dormitorios','alojamiento entero', 'seguridad en el hogar', 'dormitorio y lavandería', 'cocina y comedor',
    'maximum_guests', 'para familias', 
    'exterior', 'servicios'
]
```

Escalamos las caracteristicas que creemos más relevantes para el modelo he usado las 11 primeras del feature importance

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[features])
```

Entrenamos el modelo

```python
model = NearestNeighbors(n_neighbors=3, metric='euclidean')
model.fit(scaled_features)
```

Creamos el pickle con el modelo para streamlit

```python
with open('NearestNeighbors.pkl', 'wb') as archivo:  # wb = write binary (escribir en modo binario)
    pickle.dump(model, archivo)
```

abrimos el pickle

```python
with open('NearestNeighbors.pkl', 'rb') as archivo: #rb = read binary (leerlo)
    modelo_cargado = pickle.load(archivo)
```

Probamos el modelo con un ejemplo

```python
example_input = scaled_features[0].reshape(1, -1)
distances, indices = model.kneighbors(example_input)
```
